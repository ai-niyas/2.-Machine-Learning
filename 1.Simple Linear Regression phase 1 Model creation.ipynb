```python
#import is a statement to inherit the pandas library in python to manage data
import pandas as pd #pd is the instance of pandas
```

```python
#this is used to read the input CSV file containing data of columns and rows using the panda function read_csv function by passing the file name
#result is stored in a variable called dataset
dataset = pd.read_csv("Salary_Data.csv") 
```

```python
#to display the data stored in the variable 
dataset
```

```python
#this stage split the independent variables from the dataset by passing the field name and assigned to a variable
independent = dataset[["YearsExperience"]]
#this line list the data stored in the independent variable
independent
```

```python
#this stage split the dependent variables from the dataset by passing the field name and assigned to a variable
dependent = dataset[["Salary"]]
#this line list the data stored in the dependent variable
dependent
```

```python
# this section help to split the train and test data using sklearn package file and 
#utilize only the train_test_split function instead of importing all 
from sklearn.model_selection import train_test_split
```

```python
#train_test_split function takes the independent and dependent variable and set the train data 70%
# and the test size as 30% and random_state (to be explained later)
#and return 4 variables that hold the split data
#X_train has the training 70% train data from the independent variable and X_test has 30% of the test data from the independent variable
#y_train has the test 70% of train data from the dependent variable and y_test has 30% of the test data from the dependent variable  
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state=0)
```

```python
#display train data of independent variable
X_train
```

```python
#display test data of independent variable
X_test
```

```python
#display train data of dependent variable
y_train
```

```python
#display test data of dependent variable
y_test
```

```python
#this is to import the LinearRegression function only from Linear_model package
from sklearn.linear_model import LinearRegression
#initiate the LinearRegression functions in a container
regressor = LinearRegression()
#Build the Model with the training data using FIT function passing independent and dependent training data
regressor.fit(X_train,y_train)
```

```python
#to get the slope /weight using the coefficient function and print its value that return as an array
weight = regressor.coef_
weight
```

```python
#to get the minmum value or bias using the intercept_ function and print its value as an array
bias = regressor.intercept_
bias
```

```python
#to predict the dependent variable by passing the independent variable test data
y_pred = regressor.predict(X_test)
#print the predicted values as an array of test data
y_pred
```

```python
#to identify the r2 score for evalluation metrics using sklearn metric 
from sklearn.metrics import r2_score
#assign the r2 score from the function r2_score by passing dependent variable test data and the prediction value
r_score = r2_score(y_test,y_pred)
#print r2 score and if it's value is closer to 1 or less difference from 1 , the model is considered very good. 
#if r2_score value its closer to 0 then the model is not good
r_score
```

```python
#this package is used to create the model for deployment 
import pickle
#creating a file name to store the model with .sav file extension
filename = "finalized_model_linear.sav"
```

```python
#to write the regressor model file in write bite mode in the created file .sav file and dump it
pickle.dump(regressor,open(filename,'wb'))
```

```python
#used to load the model from the .sav file and assign it to a variable to predict the model again to check predicted value
loaded_model = pickle.load(open("finalized_model_linear.sav",'rb'))
```

```python
#used to predict the result by passing a new value other than train, test data
result = loaded_model.predict([[13]])
```
